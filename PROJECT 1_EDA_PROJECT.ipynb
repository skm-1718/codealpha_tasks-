# 1. **Which department has the highest employee attrition?**

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Department', hue='Attrition')
plt.title('Department-wise Employee Attrition')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.show()
```

### Research & Development (R&D) has the highest number of employees leaving the organization, followed by Sales. This suggests that HR should investigate workload, employee engagement, and career growth opportunities within the R&D department to reduce attrition.

```python
# Calculate attrition rate by department

department_attrition = (
    df.groupby('Department')['Attrition']
      .apply(lambda x: (x == 'Yes').mean() * 100)
      .sort_values(ascending=False)
)

print(department_attrition)
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x=department_attrition.index,
    y=department_attrition.values,
    palette='Set2'
)

plt.title('Department-wise Attrition Rate (%)')
plt.xlabel('Department')
plt.ylabel('Attrition Rate (%)')

# Add percentage labels
for i, value in enumerate(department_attrition.values):
    plt.text(i, value + 0.3, f'{value:.1f}%', ha='center')

plt.show()
```

### Although the Research & Development (R&D) department has the highest number of employees leaving, the Sales department has the highest attrition rate (%). This means employees in the Sales department are more likely to leave relative to the size of that department. Therefore, retention initiatives should prioritize the Sales department while also addressing the larger absolute number of exits in R&D.

# **Does business travel affect employee attrition?**

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,
              x='BusinessTravel',
              hue='Attrition')

plt.title('Business Travel vs Attrition')
plt.xticks(rotation=15)
plt.show()
```

## For a deeper analysis, we can calculate the attrition rate (%) within each travel category

```python
# Calculate attrition rate by Business Travel

attrition_rate = (
    df.groupby('BusinessTravel')['Attrition']
      .apply(lambda x: (x == 'Yes').mean() * 100)
      .sort_values(ascending=False)
)

print(attrition_rate)
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x=attrition_rate.index,
    y=attrition_rate.values,
    palette='viridis'
)

plt.title('Attrition Rate by Business Travel')
plt.xlabel('Business Travel')
plt.ylabel('Attrition Rate (%)')

# Display percentage values on bars
for i, value in enumerate(attrition_rate.values):
    plt.text(i, value + 0.5, f'{value:.1f}%', ha='center')

plt.show()
```

### Although Travel_Rarely has the highest number of employees leaving the company, Travel_Frequently has the highest attrition rate (%). This indicates that employees who travel frequently are more likely to leave the organization, suggesting that extensive travel may negatively impact work-life balance and employee retention.

# **Which factors are highly correlated?**

```python
plt.figure(figsize=(15,10))

corr = df.corr(numeric_only=True)

sns.heatmap(corr,
            cmap='coolwarm',
            linewidths=0.5)

plt.title('Correlation Heatmap')
plt.show()
```

### The heatmap shows relationships among numerical variables. Monthly Income has a positive relationship with Total Working Years and Job Level, while most variables have only weak linear relationships with attrition, suggesting turnover is influenced by multiple factors.
